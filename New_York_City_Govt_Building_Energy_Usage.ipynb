```python
import geopandas as gpd
import descartes
from shapely.geometry import Point, Polygon 
```

```python
street_map_NYC = gpd.read_file('geo_export_2c619fc6-39cf-4f49-84fa-a3fb7bb97faa.shp')
street_map_NYC.head()

# Extracted the shape file from the following website
# https://data.cityofnewyork.us/City-Government/Borough-Boundaries/tqmj-j8zm
```

```python
fig, ax = plt.subplots(figsize=(15,15))
street_map_NYC.plot(ax=ax, )
```

```python
address_point = [Point(xy) for xy in zip(data_NYC['Longitude'], data_NYC['Latitude'])]
address_point[:3]
```

```python
crs = {'init': 'epsg:4326'}
Geo_data_NYC = gpd.GeoDataFrame(data_NYC,
                                crs = crs, geometry=address_point)
```

```python
Geo_data_NYC.head(-10)
```

```python
fig, ax = plt.subplots(1,figsize=(15,15))
base = street_map_NYC.plot(ax=ax,color='white', edgecolor='black')
Geo_data_NYC.plot(ax=base, column='Energy', markersize=55, cmap='cool', alpha=0.9, legend=True, label='Energy')
plt.legend()
plt.title('Energy used at different Government building in New York city, FIY2015, in MMBTU')
plt.savefig('Energy_Usage_Government_Building_NYC.png', bbox_inches ='tight')

# This website was very helpful to combine (map and the co-ordinate system)  and color assignment to data points 
# https://www.datacamp.com/community/tutorials/geospatial-data-python
# MMBTU means Million British Thermal Units ( It is also called as MBTU, BTU. )
# 1 BTU : 1.06 Joule
# A BTU is a measure of the energy content in fuel, and is used in the power, 
# steam generation, heating and air conditioning industries. 
```
